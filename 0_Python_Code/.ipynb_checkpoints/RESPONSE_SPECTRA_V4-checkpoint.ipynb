```python
initial_num = 1
final_num = 2
#T = np.zeros((1000,final_num))
df_disp = pd.DataFrame()
df_vel = pd.DataFrame()
df_acc = pd.DataFrame()


df_disp = pd.DataFrame(index=range(1000),columns=range(0))
df_vel = pd.DataFrame(index=range(1000),columns=range(0))
df_acc = pd.DataFrame(index=range(1000),columns=range(0))


fig = plt.figure()                           
#fig = plt.figure(figsize=(fig_size_x, fig_size_y))                       # size of ffigure
ax = fig.gca()
plt.title("Sudo Velocity Spectrum")      
plt.xlabel("Period (sec)")                               
plt.ylabel("Spectral Velocity")                                              
plt.grid()                                                                # X-AXIS lower & upper limit
#plt.xlim(x_lower_limit,x_upper_limit)                                     # Y-AXIS Lower & Upper Limit
#plt.ylim(y_lower_limit, y_upper_limit)                                    # 
#resolution_value = image_resolution   

 
while initial_num <= final_num   :
    
# Import xls file    
    el = pd.read_excel('ELC%d.xls'%initial_num)
    t = el.iloc[:,0]
    E = np.zeros((len(el.iloc[:,1]),1))
    E[:,0] =el.iloc[:,1]
    
    
# INPUT DATA
    xi = 0.1
    m = 1
    dt = t[1] - t[0]
    T = np.zeros((1000,1))
    T[:,0] = np.linspace(0.01,10,1000)[:]
    omega_n = 2*np.pi/T
    k = m*omega_n*omega_n
    
#SOLUTION
    import numpy as np
    import scipy.linalg as la
    
    
    u0 = np.zeros((len(omega_n),1))
    for j in range(len(omega_n)):
        wn = omega_n[j,0]
        A = np.array([[0,1],[-wn**2, -2*xi*wn]])
        D, V = la.eig(A)
        ep = np.array([[np.exp(D[0]*dt),0],[0,np.exp(D[1]*dt)]])
        Ad = V.dot(ep).dot(np.linalg.inv(V))
        Bd = np.linalg.inv(A).dot(Ad-np.eye((len(A))))
        z = [[0],[0]]
        d = np.zeros((len(E),1))
        v = np.zeros((len(E),1))
    
        for i in range(len(E)):
            z = np.real(Ad).dot(z) + np.real(Bd).dot([[0],[-E[i]]])    #[-E[i]*386.09]])
            d[i] = z[0,0]
            v[i] = z[1,0]
        u0[j] = max(abs(d))
        
        #plt.plot(T,u0)
        #plt.show()
    
    # Response Spectral
    u0
    
    V = omega_n*u0
    A = omega_n*omega_n*u0
    
#    v = V.tolist()
    
    df_disp['data_%d'%initial_num] = u0
    df_vel['data_%d'%initial_num] = V
    df_acc['data_%d'%initial_num] = A
    
    #Vel = V.tolist()
    #vel.append(Vel)
    
    #Acc =A.tolist()
    #acc.append(Acc)
    
    #Disp = u0.tolist()
    #disp.append(Disp)
    
    

    plt.plot(T,V)

    initial_num = initial_num + 1

#plt.show()
#fig.savefig("Velcoity_Spectra.png")

df_disp['mean'] = df_disp.mean(axis=1)
df_vel['mean'] = df_vel.mean(axis=1)
df_acc['mean'] = df_acc.mean(axis=1)


df_disp.to_csv(r'Displacement Spectra.csv')
df_vel.to_csv(r'Velocity Spectra.csv')
df_acc.to_csv(r'Acceleration Spectra.csv')

mean_1 = df_vel['mean']

plt.plot(T,mean_1)

plt.show()
fig.savefig("Velcoity_Spectra.png")
```
